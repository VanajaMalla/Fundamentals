```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,5*np.pi,0.1)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
y4=np.square(x)
y5=np.exp(x)
plt.figure(1)

plt.subplot(221)
plt.plot(x,y1,label='sin(t)')
plt.plot(x,y2,label='cos(t)')  
plt.xlabel("t")
plt.grid()
plt.legend()

plt.subplot(222)
plt.plot(x,y3,label='tan(t)')
plt.legend()

plt.subplot(223)
plt.plot(x,y4,label='x^2')
plt.legend()

plt.subplot(224)
plt.plot(x,y5,label='e^x')
plt.legend()
plt.show()
```
